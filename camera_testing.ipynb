```python
from jetbot import Camera,bgr8_to_jpeg
import matplotlib.pyplot as plt
import numpy as np
import cv2
import time
```

```python
# import pandas as pd
```

```python
from pathlib import Path
```

```python
p = "images"
p = Path(p)
p.mkdir(exist_ok=True)
```

```python
# https://www.scivision.co/numpy-image-bgr-to-rgb/
def convert_bgr_to_rgb(x): return x[...,::-1]
```

```python
c = Camera()
c.fps = 2
c.start()
```

```python
img = convert_bgr_to_rgb(c.value).astype('uint8')
```

```python
cv2.imwrite(str(p / "test.png"),img) 
```

```python
for i in range(100):
    img = convert_bgr_to_rgb(c.value)
    r=cv2.imwrite(str(p / f"test{i}.png"),img) 
    print(f"image num {i}. saved ok {r}")    
    time.sleep(0.5)
```

```python
img = convert_bgr_to_rgb(c.value).astype('uint8')
plt.figure(figsize = (8,8))
plt.imshow(img)
```

```python
c.stop()
```
